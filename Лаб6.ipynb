```python
# Задание task_06_02_01.

def foo(a):
    """!!!

    Сложность: O(N**2).
    """
    for i in range(len(a), 0, -1):                #O(N)
        for j in range(1, i):                     #O(N)
            if a[j-1] > a[j]:                     #O(1)
                a[j-1], a[j] = a[j], a[j-1]       #O(1)
    return a                                      #O(1)


a = [1, 2, 3, 4, 2, 1, 3, 4, 3, 5, -2, -2, 2, -2, 65, 2, 2]
foo(a)
```

```python
# Задание task_06_02_03.

def foo(s):
    """!!!

    Параметры:
        - s (str): строка.

    Сложность: O(N**2).
    """
    val = 0                     #O(1)
    for c in s:                 #O(N)
        if c.isdigit():         #O(N)
            val += int(c)       #O(1)
    return val                  #O(1)
```

```python
# Задание task_06_02_04.

def foo(n):
    """!!!

    Параметры:
        - n (int): число.

    Сложность: O(N).
    """
    res = []                                   #O(N)
    for i in range(1, n + 1):                  #O(N)
        divisors = 0                           #O(1)
        j = 2                                  #O(1)
        while j < i and divisors == 0:         #O(1)
            if i % j == 0:                     #O(1)
                divisors += 1                  #O(1)
            j += 1                             #O(1)

        if divisors == 0:                      #O(1)
            res.append(i)                      #O(1)

    return res                                 #O(1)
```
